# Task
Scrape movie titles, Kinopoisk ratings, and IMDb ratings from the website "https://kinogo.am/" across multiple pages (up to 50 pages or 500 movies, whichever comes first) and save the collected data into a CSV file named `movies_data_full.csv`. If a rating is missing for a movie, record it as `None`.

## Инициализация списков данных

### Subtask:
Создать пустые списки для хранения заголовков фильмов, рейтингов Кинопоиска и рейтингов IMDb.


**Reasoning**:
The user wants to initialize three empty lists to store movie titles, Kinopoisk ratings, and IMDb ratings. This code block will create those lists.



```python
all_titles = []
all_kp_ratings = []
all_imdb_ratings = []

print("Empty lists 'all_titles', 'all_kp_ratings', and 'all_imdb_ratings' created.")
```

## Цикл по страницам

### Subtask:
Просматривать страницы сайта до тех пор, пока не будет собрано около 10 000 фильмов (или до последней доступной страницы, если фильмов меньше). Для каждой страницы: создать URL, отправить HTTP-запрос, получить HTML-содержимое.


**Reasoning**:
The subtask requires iterating through website pages, constructing URLs, sending HTTP requests, and parsing HTML content. This code block will implement the loop structure, handling URL generation, making requests, parsing the response with BeautifulSoup, and storing the parsed content for future processing.



```python
page_num = 1
scraped_movies = []
max_pages = 50
max_movies = 500

all_soups = []

while page_num <= max_pages and len(scraped_movies) < max_movies:
    if page_num == 1:
        page_url = url
    else:
        page_url = f"{url}/page/{page_num}/"

    print(f"Scraping page {page_num}: {page_url}")

    response = requests.get(page_url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html')
        all_soups.append(soup) # Store the soup object


    else:
        print(f"Failed to retrieve page {page_num}. Status code: {response.status_code}")
        break

    page_num += 1

print(f"Finished iterating through pages. Collected {len(all_soups)} soup objects.")
```

**Reasoning**:
The previous step successfully iterated through the specified number of pages, fetched the HTML content, and parsed it into `BeautifulSoup` objects, storing them in `all_soups`. The next logical step is to extract the required information (titles, KP ratings, IMDb ratings) from each of these `soup` objects and append them to the respective initialized lists (`all_titles`, `all_kp_ratings`, `all_imdb_ratings`). This will complete the data collection for all pages.



```python
for soup_obj in all_soups:
    titles_elements = soup_obj.find_all('div', class_='shortstory__title')
    kp_elements = soup_obj.find_all('span', class_='kp')
    imdb_elements = soup_obj.find_all('span', class_='imdb')

    # Extract text and append to the all_lists
    for title_elem in titles_elements:
        all_titles.append(title_elem.text.strip())

    for kp_elem in kp_elements:
        all_kp_ratings.append(kp_elem.text.strip())

    for imdb_elem in imdb_elements:
        all_imdb_ratings.append(imdb_elem.text.strip())

    # Update scraped_movies for the loop condition in the next run
    # This assumes each shortstory__title corresponds to one movie for counting purposes
    # The previous `scraped_movies` was just an empty list. Now we will actually count the movies.
    scraped_movies.extend(titles_elements)

print(f"Total titles collected: {len(all_titles)}")
print(f"Total KP ratings collected: {len(all_kp_ratings)}")
print(f"Total IMDb ratings collected: {len(all_imdb_ratings)}")
print(f"Total movies counted for loop condition: {len(scraped_movies)}")
```

**Reasoning**:
The previous extraction process resulted in lists of different lengths for titles, KP ratings, and IMDb ratings because missing ratings were skipped. To ensure that each movie corresponds to its ratings (or `None` if a rating is missing), it's necessary to iterate through each 'shortstory' element, extract the title, and then attempt to extract its corresponding KP and IMDb ratings, assigning `None` if they are not found. This will synchronize the lengths of the lists and correctly represent missing data.



```python
all_titles = []
all_kp_ratings = []
all_imdb_ratings = []

for soup_obj in all_soups:
    shortstories = soup_obj.find_all('div', class_='shortstory')
    for story in shortstories:
        title_elem = story.find('div', class_='shortstory__title')
        if title_elem:
            all_titles.append(title_elem.text.strip())
        else:
            all_titles.append(None) # Should not happen if shortstory is correctly identified

        kp_elem = story.find('span', class_='kp')
        if kp_elem:
            all_kp_ratings.append(kp_elem.text.strip())
        else:
            all_kp_ratings.append(None)

        imdb_elem = story.find('span', class_='imdb')
        if imdb_elem:
            all_imdb_ratings.append(imdb_elem.text.strip())
        else:
            all_imdb_ratings.append(None)

print(f"Total titles collected: {len(all_titles)}")
print(f"Total KP ratings collected: {len(all_kp_ratings)}")
print(f"Total IMDb ratings collected: {len(all_imdb_ratings)}")
```

**Reasoning**:
The collected movie titles, Kinopoisk ratings, and IMDb ratings are now in three lists of equal length. The next step is to combine these lists into a pandas DataFrame and then save this DataFrame to a CSV file as specified in the main task, `movies_data_full.csv`.



```python
df_full = pd.DataFrame({
    'Title': all_titles,
    'KP Rating': all_kp_ratings,
    'IMDb Rating': all_imdb_ratings
})

print(df_full.head())

df_full.to_csv('movies_data_full.csv', index=False)
print("Data saved to 'movies_data_full.csv'")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
import re   # <-- added for year extraction

all_movies_data = []
print("Empty list 'all_movies_data' created.")

base_url = "https://kinogo.am"

for page_num in range(1, 101):

    if page_num == 1:
        url = base_url
    else:
        url = f"{base_url}/page/{page_num}/"

    try:
        response = requests.get(url, timeout=10)
        response.raise_for_status()

        soup = BeautifulSoup(response.text, "html.parser")

        shortstories = soup.find_all("div", class_="shortstory")

        if not shortstories:
            print(f"No 'shortstory' blocks found on page {page_num}. Stopping.")
            break

        for story in shortstories:

            # ---- Title ----
            title_block = story.find("div", class_="shortstory__title")
            title_h2 = title_block.find("h2") if title_block else None
            title = title_h2.text.strip() if title_h2 else None

            # ---- KP rating ----
            kp_block = story.find("span", class_="kp")
            kp = kp_block.text.replace("KP ", "").strip() if kp_block else None

            # ---- IMDb rating ----
            imdb_block = story.find("span", class_="imdb")
            imdb = imdb_block.text.replace("IMDB ", "").strip() if imdb_block else None

            # ---- YEAR extraction ----
            info_block = story.find("div", class_="shortstory__info")
            info_text = info_block.text.strip() if info_block else ""

            # Extract first 4-digit number (likely year)
            year_match = re.search(r"\b(19|20)\d{2}\b", info_text)
            year = year_match.group(0) if year_match else None

            all_movies_data.append({
                "Название фильма": title,
                "Год": year,
                "Рейтинг KP": kp,
                "Рейтинг IMDB": imdb
            })

        print(f"Scraped page {page_num}. Total movies: {len(all_movies_data)}")

    except requests.exceptions.RequestException as e:
        print(f"Request error on page {page_num}: {e}")
        break
    except Exception as e:
        print(f"Unexpected error on page {page_num}: {e}")

    time.sleep(0.25)

print(f"\nFinished scraping. Total movies collected: {len(all_movies_data)}")

Adilzhan = pd.DataFrame(all_movies_data)

print("\nFirst 5 rows:")
display(Adilzhan.head())

print("\nLast 5 rows:")
display(Adilzhan.tail())

print(f"\nShape of DataFrame: {Adilzhan.shape}")
```
